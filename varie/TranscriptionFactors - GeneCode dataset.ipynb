```python
import pandas as pd
import numpy as np

df = pd.read_csv("varie/attribute_set_library_crisp.txt", header=None, sep='\n')
df = df[0].str.split('\t', expand=True)
df = df.drop(columns=[1])
df = df.rename(columns={0: "GeneCode"})

df
```

```python
transcription_factors_columns = list(range(2, len(df.columns) + 1))

transcription_factors = set()

for _, row in df.iterrows():

  for column_index in transcription_factors_columns:
    if row[column_index] is not None:
      transcription_factors.add(row[column_index])

transcription_factors = list(transcription_factors)
transcription_factors.sort()
```

```python
def temp(x):
  t = []
  for el in x:
    if el is not None:
      t.append(el)
  return t
df = df.set_index('GeneCode')
val = map(temp,df.values)
df_buono = pd.DataFrame({'GeneCode':df.index,'TF':list(val)})
```

```python
df_buono.values
```

```python
def micro_encoder(l):
  vec = []
  for el in transcription_factors: 
    if el in l[1]:
      vec.append(1)
    else:
      vec.append(0)
  return vec
```

```python
df_finale = pd.DataFrame({'GeneCode':df.index,'TF':list(map(micro_encoder,df_buono.values))})
df_finale
```
